```python
import os
import shutil
import datetime
import schedule
import time

source_dir = "/content/drive/MyDrive/Temp"
destination = "/content/drive/MyDrive/Backup"

def copy_folder_to_directory(source, dest):
  today = datetime.date.today()
  dest_dir = os.path.join(dest, str(today))

  try:
    shutil.copytree(source, dest_dir)
    print(f"Folder copied to: {dest_dir}")
  except FileExistsError:
    print(f"Folder already exists in: {dest_dir}")

def l():
  copy_folder_to_directory(source_dir, destination)

schedule.every().day.at("17:30").do(l)

while True:
  schedule.run_pending()
  time.sleep(1)
```
